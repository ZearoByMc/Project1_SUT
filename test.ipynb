```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from skimage import data, img_as_float

# ฟังก์ชันช่วยแปลงภาพเข้าสู่ Frequency Domain (FFT)
def to_freq_domain(img):
    f = np.fft.fft2(img)
    fshift = np.fft.fftshift(f)
    return fshift

# ฟังก์ชันช่วยแปลงกลับสู่ Spatial Domain (Inverse FFT)
def to_spatial_domain(fshift):
    f_ishift = np.fft.ifftshift(fshift)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)
    return img_back

# โหลดภาพตัวอย่าง (ภาพตากล้อง) และแปลงเป็นขาวดำ
original_img = img_as_float(data.camera())
rows, cols = original_img.shape
crow, ccol = rows//2, cols//2 # จุดกึ่งกลางภาพ
```
